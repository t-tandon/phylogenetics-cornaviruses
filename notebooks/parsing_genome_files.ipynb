```python
#changing header

in_file = "sequences.fasta"
out_file = "sequences.fasta.cleaned"

with open(in_file) as infile, open(out_file, "w") as outfile:
    for line in infile:
        if line.startswith(">"):
            header = line[1:].strip()
            header = header.replace("complete genome", "")
            header = header.replace("|", "_").replace(" ", "_").replace(",", "")
            outfile.write(f">{header}\n")
        else:
            outfile.write(line)
```

```python
#histogram for bp 

import matplotlib.pyplot as plt

lengths = []
with open("sequences.fasta.cleaned", "r") as file: 
    length = 0 
    for line in file: 
        if line.startswith(">"): 
            if length>0: 
                lengths.append(length) 
            length = 0 #reset
        else: 
            length += len(line.strip())
    
plt.hist(lengths, bins = 20, color = "pink")
plt.xlabel("genome length(bp)")
plt.ylabel("number of genomes")
plt.title("histogram of genome lengths") 
plt.show()

```
